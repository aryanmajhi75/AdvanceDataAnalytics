```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read the image
image_path = '/content/drive/MyDrive/Trimester 4/anupam.jpg'  # Replace 'your_image_path.jpg' with the path to your image
original_image = cv2.imread(image_path)
```

```python
# modify contrast and Brightness
brightness = 30
contrast = 1.2
adjusted_image = np.clip(original_image * contrast + brightness, 0, 255).astype(np.uint8)
```

```python

# Convert to grayscale
grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)
```

```python

# Calculate histograms
hist_original = cv2.calcHist([original_image], [0], None, [256], [0, 256])
hist_adjusted = cv2.calcHist([adjusted_image], [0], None, [256], [0, 256])
hist_grayscale = cv2.calcHist([grayscale_image], [0], None, [256], [0, 256])
```

```python

# Plot histograms
plt.figure(figsize=(10, 6))

plt.subplot(3, 2, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(3, 2, 2)
plt.plot(hist_original, color='blue')
plt.title('Histogram (Original)')

plt.subplot(3, 2, 3)
plt.imshow(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
plt.title('Adjusted Image')
plt.axis('off')

plt.subplot(3, 2, 4)
plt.plot(hist_adjusted, color='green')
plt.title('Histogram (Adjusted)')

plt.subplot(3, 2, 5)
plt.imshow(grayscale_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.subplot(3, 2, 6)
plt.plot(hist_grayscale, color='red')
plt.title('Histogram (Grayscale)')

plt.tight_layout()
plt.show()
```
