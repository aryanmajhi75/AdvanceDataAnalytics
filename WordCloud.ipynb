## Word Cloud after removing Stopwords

```python
stopwords=set(stopwords.words('english'))
```

```python
word1=word_tokenize(para1)
```

```python
filtered1=[]

for w in word1:
    if w not in stopwords:
        filtered1.append(w)
```

```python
wordcloud3 = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50,
                          random_state=42
                         ).generate(str(filtered1))

print(wordcloud1)
fig = plt.figure(1)
plt.imshow(wordcloud3)
plt.axis('off')
plt.show()
# fig.savefig("word1.png", dpi=900)
```

```python
word2=word_tokenize(para2)
```

## WordCloud after Stemming

```python
ps = PorterStemmer()
```

```python
poster_stem1 = []

for w in word1:
    poster_stem1.append(ps.stem(w))
    print(w, " : ", ps.stem(w))

print("\n [poster_stem1] \n",poster_stem1)
print("-----------------------------------------------------")
print("\nLength or rather number of words ",len(poster_stem1))
```

```python
wordcloud5 = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50,
                          random_state=42
                         ).generate(str(poster_stem1))

print(wordcloud5)
fig = plt.figure(1)
plt.imshow(wordcloud5)
plt.axis('off')
plt.show()
```

```python
poster_stem2 = []

for w in word2:
    poster_stem2.append(ps.stem(w))
    print(w, " : ", ps.stem(w))

# print("\n\n",poster_stem1)
print("\nLength or rather number of words ",len(poster_stem2))
```

```python
wordcloud6 = WordCloud(
                          background_color='black',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50,
                          random_state=42
                         ).generate(str(poster_stem2))

print(wordcloud6)
fig = plt.figure(1)
plt.imshow(wordcloud6)
plt.axis('off')
plt.show()
```

## WordCloud after Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemma1=[]

for words in word1:
  lemma1.append(lemmatizer.lemmatize(words))
  print(words," : ",lemmatizer.lemmatize(words))
print("\n\n",lemma1)
```

```python
wordcloud7 = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50,
                          random_state=42
                         ).generate(str(lemma1))

print(wordcloud7)
fig = plt.figure(1)
plt.imshow(wordcloud7)
plt.axis('off')
plt.show()
```

```python
lemma2=[]

for words in word2:
  lemma2.append(lemmatizer.lemmatize(words))
  print(words," : ",lemmatizer.lemmatize(words))
print("\n\n",lemma2)
```

```python
wordcloud8 = WordCloud(
                          background_color='black',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50,
                          random_state=42
                         ).generate(str(lemma2))

print(wordcloud8)
fig = plt.figure(1)
plt.imshow(wordcloud8)
plt.axis('off')
plt.show()
```
